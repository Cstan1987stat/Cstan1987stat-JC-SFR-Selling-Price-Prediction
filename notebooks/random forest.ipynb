```python
# The below lines of code import the pandas library and the display function.
import pandas as pd 
from IPython.display import display

# The below lines of code read in the train_df and test_df CSV files.
train = pd.read_csv('train_df.csv')
test = pd.read_csv('test_df.csv')

# The below lines of code display the first five rows of the train data frame.
print('First five rows of train data frame:')
display(train.head())


print()
print()
# The below lines of code display the first five rows of the test data frame.
print('First five rows of test data frame:')
display(test.head())

# The below lines of code delete the previous index column in both the train and test data frames. 
del train['Unnamed: 0']
del test['Unnamed: 0']


print()
print()
# The below lines of code display the first five rows of the train data frame.
print('First five rows of train data frame after removing unnecesary column:')
display(train.head())

print()
print()
# The below lines of code display the first five rows of the test data frame.
print('First five rows of test data frame after removing unnecesary column:')
display(test.head())


from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# The below lines of code assign the Price column from the train and test data frames as the target variables.
train_target = train['Price']
test_target = test['Price']

# The below lines of code delete the Price column from the train and test data frames. 
del train['Price']
del test['Price']


# The below lines of code create a transformation pipeline that applies a logarithmic transformation followed by a standard scaling.
pipe1 = Pipeline([
    ('log', FunctionTransformer(func=np.log1p)),
    ('scaler', StandardScaler()),
    ])

# The below line of code retrieves the column names from the train data frame.
c_names = train.columns


# The below line of code fits the pipe1 pipeline with the training data, and then transforms the train data with the fitted pipeline.
train = pipe1.fit_transform(train)
# The below line of code transforms the test data with the fitted pipeline.
test = pipe1.transform(test)

# The below lines of code create new train and test dataframes using the original column names. 
train = pd.DataFrame(train, columns=c_names)
test = pd.DataFrame(test, columns=c_names)

# The below line of code creates a logrithmically transformed target variable.
train_target_log = np.log(train_target)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

import seaborn as sns
import matplotlib.pyplot as plt
```

## Random Forest Regressor Parameters

### n_estimators:

Accoring to the RandomForestRegressor documentation on scikit-learn, this is simple of the number of decision trees that will be used in the model [1].

```python
# The below lines of code create a train_df and test_df data frames to store C values and corresponding MAEs.
train_df = pd.DataFrame(columns=['n_estimators','mae'])
test_df = pd.DataFrame(columns=['n_estimators','mae'])

# The below line of code creates a for loop that goes from 1 to 200, increasing in 10 unit increments.
for i in np.arange(1, 200, 10):    
    '''
    - Creating a RandomForestRegressor with given value of n_estimators.
    - Fitting the RandomForestRegressor, calculating the mean absolute error for the training and testing data.
    - Updating the train_df and test_df when the mean absolute error and given value for n_estimators.
    '''
    rfr = RandomForestRegressor(random_state=42, n_estimators=i)
    rfr.fit(train, train_target_log)
    pred_target_log = rfr.predict(train)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(train_target, pred_target)
    train_df.loc[len(train_df.index)] = [i, mae]
    pred_target_log = rfr.predict(test)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(test_target, pred_target)
    test_df.loc[len(test_df.index)] = [i, mae]

# The following lines of code plot the max_depth on the x-axis for training and testing data, with the mae for training and testing data on the y-axis.    
sns.lineplot(x = train_df['n_estimators'], y = train_df['mae'], color='#8B0000', linestyle='dashed')
sns.lineplot(x = test_df['n_estimators'], y = test_df['mae'], color='#8B0000', linestyle='dotted')
plt.title('Training data MAE (dashed) and testing data MAE (dotted)') # Creating a title for the graph.
plt.xlabel('n_estimators')  # Creating an x-axis label.
plt.show()
```

It's immediately clear that overfitting will be a significant issue. The MAE for nearly all Random Forest Regressor models on the training data is below 30,000, while the lowest MAE on the testing data is around 75000. Around 10 n_estimators, is where the MAE for both the training and testing data hover at around the same value.

### max_depth

This parameter refers to how many layers each decision tree will have [1]. 

```python
# The below lines of code create a train_df and test_df data frames to store C values and corresponding MAEs.
train_df = pd.DataFrame(columns=['max_depth','mae'])
test_df = pd.DataFrame(columns=['max_depth','mae'])

# The below line of code creates a for loop that goes from 1 to 200, increasing in 10 unit increments.
for i in np.arange(1, 200, 10):    
    '''
    - Creating a RandomForestRegressor with the given value of max_depth.
    - Fitting the RandomForestRegressor, calculating the mean absolute error for the training and testing data.
    - Updating the train_df and test_df when the mean absolute error and given value for max_depth.
    '''
    rfr = RandomForestRegressor(random_state=42, n_estimators=i)
    rfr.fit(train, train_target_log)
    pred_target_log = rfr.predict(train)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(train_target, pred_target)
    train_df.loc[len(train_df.index)] = [i, mae]
    pred_target_log = rfr.predict(test)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(test_target, pred_target)
    test_df.loc[len(test_df.index)] = [i, mae]  

# The following lines of code plot the max_depth on the x-axis for training and testing data, with the mae for training and testing data on the y-axis.
sns.lineplot(x = train_df['max_depth'], y = train_df['mae'], color='#8B0000', linestyle='dashed')
sns.lineplot(x = test_df['max_depth'], y = test_df['mae'], color='#8B0000', linestyle='dotted')
plt.title('Training data MAE (dashed) and testing data MAE (dotted)') # Creating a title for the graph.
plt.xlabel('max_depth')   # Creating an x-axis label.   
plt.show()
```

Once again, overfitting does seem to be an inevitable issue. The lower the max_depth, the more similar the MAE for the random forest regressor on the training and testing data. Overall, it becomes clear that at a max_depth of around 10, the MAE is constant for both the training and testing data.

### min_samples split

This parameter refers to the minimum number of data points that are required to split up a node [1].

```python
# The below lines of code create a train_df and test_df data frames to store C values and corresponding MAEs.
train_df = pd.DataFrame(columns=['min_samples_split','mae'])
test_df = pd.DataFrame(columns=['min_samples_split','mae'])

# The below line of code creates a for loop that goes from 2 to 200, increasing in 8 unit increments.
for i in np.arange(2, 100, 8):
    '''
    - Creating a RandomForestRegressor with the given value of min_samples_split.
    - Fitting the RandomForestRegressor, calculating the mean absolute error for the training and testing data.
    - Updating the train_df and test_df when the mean absolute error and given value for min_samples_split.
    '''        
    rfr = RandomForestRegressor(random_state=42, min_samples_split=i)
    rfr.fit(train, train_target_log)
    pred_target_log = rfr.predict(train)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(train_target, pred_target)
    train_df.loc[len(train_df.index)] = [i, mae]
    pred_target_log = rfr.predict(test)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(test_target, pred_target)
    test_df.loc[len(test_df.index)] = [i, mae]

# The following lines of code plot the min_sample_split on the x-axis for training and testing data, with the mae for training and testing data on the y-axis.
sns.lineplot(x = train_df['min_samples_split'], y = train_df['mae'], color='#8B0000', linestyle='dashed')
sns.lineplot(x = test_df['min_samples_split'], y = test_df['mae'], color='#8B0000', linestyle='dotted')
plt.title('Training data MAE (dashed) and testing data MAE (dotted)') # Creating a title for the graph.
plt.xlabel('min_samples_split')   # Creating an x-axis label.   
plt.show()
```

As the min_samples_split increases, the MAE for the training data dramatically increases. The MAE for the testing data also increases, but not as drastically.

### min_samples_leaf

This parameter refers to the minimum number of data points required at a leaf node [1]. If it’s set to 1, and a potential split would leave one of the resulting nodes empty, the split won’t be allowed because the minimum samples per leaf is 1.

```python
# The below lines of code create a train_df and test_df data frames to store min_samples_leaf values and corresponding MAEs.
train_df = pd.DataFrame(columns=['min_samples_leaf','mae'])
test_df = pd.DataFrame(columns=['min_samples_leaf','mae'])

# The following line of code creates a for loop that goes from 1 to 100, increasing in 8 unit increments
for i in np.arange(1, 100, 8):    
    '''
    - Creating a RandomForestRegressor with the given value of min_samples_lead.
    - Fitting the RandomForestRegressor, calculating the mean absolute error for the training and testing data.
    - Updating the train_df and test_df when the mean absolute error and the given value for min_samples_leaf.
    '''        
    rfr = RandomForestRegressor(random_state=42, min_samples_leaf=i)
    rfr.fit(train, train_target_log)
    pred_target_log = rfr.predict(train)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(train_target, pred_target)
    train_df.loc[len(train_df.index)] = [i, mae]
    pred_target_log = rfr.predict(test)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(test_target, pred_target)
    test_df.loc[len(test_df.index)] = [i, mae]



# The following lines of code plot the min_samples_leaf on the x-axis for training and testing data, with the mae for training and testing data on the y-axis.
sns.lineplot(x = train_df['min_samples_leaf'], y = train_df['mae'], color='#8B0000', linestyle='dashed')
sns.lineplot(x = test_df['min_samples_leaf'], y = test_df['mae'], color='#8B0000', linestyle='dotted')
plt.title('Training data MAE (dashed) and testing data MAE (dotted)') # Creating a title for the graph.
plt.xlabel('min_samples_leaf')   # Creating an x-axis label.   
plt.show()
```

On the left side of the graph, there’s clear overfitting. As the min_samples_leaf parameter increases, the MAE for training data rises significantly and eventually surpasses the MAE for testing data, suggesting it could act as a generalization parameter. However, it doesn’t work as expected, as the MAE for testing data never decreases, leaving the usefulness of this parameter unclear.

### max_features

This parameter referes to the number of features that the regressor can consider when looking for the best split [1].

```python
# The below lines of code create a train_df and test_df data frames to store max_features values and corresponding MAEs.
train_df = pd.DataFrame(columns=['max_features','mae'])
test_df = pd.DataFrame(columns=['max_features','mae'])

# The following line of code creates a for loop that goes from 1 to 49, increasing in 1 unit increments.
for i in np.arange(1, 49, 1):    
    '''
    - Creating a RandomForestRegressor with the given value of max_features.
    - Fitting the RandomForestRegressor, calculating the mean absolute error for the training and testing data.
    - Updating the train_df and test_df when the mean absolute error and the given value for max_features.
    '''      
    rfr = RandomForestRegressor(random_state=42, max_features=i)
    rfr.fit(train, train_target_log)
    pred_target_log = rfr.predict(train)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(train_target, pred_target)
    train_df.loc[len(train_df.index)] = [i, mae]
    pred_target_log = rfr.predict(test)
    pred_target = np.exp(pred_target_log)
    mae = mean_absolute_error(test_target, pred_target)
    test_df.loc[len(test_df.index)] = [i, mae]

# The following lines of code plot the max_features on the x-axis for training and testing data, with the mae for training and testing data on the y-axis.
sns.lineplot(x = train_df['max_features'], y = train_df['mae'], color='#8B0000', linestyle='dashed')
sns.lineplot(x = test_df['max_features'], y = test_df['mae'], color='#8B0000', linestyle='dotted')
plt.title('Training data MAE (dashed) and testing data MAE (dotted)') # Creating a title for the graph.
plt.xlabel('max_features')   # Creating an x-axis label.   
plt.show()
```

Right away, it is clear that no matter how big or small the max_feature parameter is set to, overfitting will always be an major issue (it seems that there is always a 30,000-40,000 difference between the MAE for training data compared to testing data).

## Semi Grid Search

```python
# The following line of code imports the RandomizedSearchCV function.
from sklearn.model_selection import RandomizedSearchCV
```

```python
# The following lines of code create a dictionary with different hyperparameter values.
param_grid = {
    'n_estimators': np.arange(25,80,5),
    'max_depth': np.arange(20,65,5),
    'min_samples_split': np.arange(2,7,1),
    'min_samples_leaf': np.arange(1,4,1),
    'max_features': np.arange(10,24,1),
}

# The below line of code specifies an intial Random Forest Regressor.
rfr = RandomForestRegressor(random_state=42)

# The following initalizes a RandomizedSearchCV model with the Random Forest Regressor, using param_grid for hyparameters, neg_mean_absolute_error as the scoring metric, 4-fold cross-validation, and a random state for reproducibility.
grid_search = RandomizedSearchCV(rfr, 
                                 param_distributions=param_grid,
                                 scoring='neg_mean_absolute_error',
                                 cv=4,
                                 return_train_score=True,
                                 random_state=22
                                 )

# The below line of code fits the grid_search with train and train_target.
grid_search.fit(train, train_target)
```

```python
# The below line of code isolates the best estimator from randomized search as rfr.
rfr = grid_search.best_estimator_
print(rfr)
print()
# The following lines of code fit a Random Forest Regressor and calculates the mean absolute the mean absolute error of the model on the training and testing data.
rfr.fit(train, train_target_log)
pred_target_log = rfr.predict(train)
pred_target = np.exp(pred_target_log)
print('Training data MAE:', np.round(mean_absolute_error(train_target, pred_target),2))
pred_target_log = rfr.predict(test)
pred_target = np.exp(pred_target_log)
print('Testing data MAE:', np.round(mean_absolute_error(test_target, pred_target),2))
```

The training data MAE for this Random Forest Regressor is 35804.05. What we can take away is that when tested with the training data, our model is off by $35804.95. 
The testing data MAE for this Random Forest Regressor is 70116.96. What we can take away is that when tested with the testing data, our model is off by $70116.96.

While the training data score is fairly decent, the increase in MAE from training to testing data is alarming.
