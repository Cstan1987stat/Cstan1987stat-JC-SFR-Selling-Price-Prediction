```python
# The below lines of code import the pandas library and the display function.
import pandas as pd 
from IPython.display import display

# The below lines of code read in the train_df and test_df CSV files.
train = pd.read_csv('train_df.csv')
test = pd.read_csv('test_df.csv')

# The below lines of code display the first five rows of the train data frame.
print('First five rows of train data frame:')
display(train.head())


print()
print()
# The below lines of code display the first five rows of the test data frame.
print('First five rows of test data frame:')
display(test.head())

# The below lines of code delete the previous index column in both the train and test data frames. 
del train['Unnamed: 0']
del test['Unnamed: 0']


print()
print()
# The below lines of code display the first five rows of the train data frame.
print('First five rows of train data frame after removing unnecesary column:')
display(train.head())

print()
print()
# The below lines of code display the first five rows of the test data frame.
print('First five rows of test data frame after removing unnecesary column:')
display(test.head())
```

## Loading Functions/Libraries

```python
# The below line of code imports the SVR function.
from sklearn.svm import SVR
# The below line of code imports the mean absolute error function.
from sklearn.metrics import mean_absolute_error

# Imporing packages/libraries for transformations.
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# The below lines of code import data visualization libraries.
import seaborn as sns
import matplotlib.pyplot as plt
```

## Prework before model fitting

```python
# The below lines of code assign the Price column from the train and test data frames as the target variables.
train_target = train['Price']
test_target = test['Price']

# The below lines of code delete the Price column from the train and test data frames. 
del train['Price']
del test['Price']

# The below lines of code create a transformation pipeline that applies a logarithmic transformation followed by a standard scaling.
pipe1 = Pipeline([
    ('log', FunctionTransformer(func=np.log1p)),
    ('scaler', StandardScaler()),
    ])

# The below line of code retrieves the column names from the train data frame.
c_names = train.columns


# The below line of code fits the pipe1 pipeline with the training data, and then transforms the train data with the fitted pipeline.
train = pipe1.fit_transform(train)
# The below line of code transforms the test data with the fitted pipeline.
test = pipe1.transform(test)

# The below lines of code create new train and test dataframes using the original column names. 
train = pd.DataFrame(train, columns=c_names)
test = pd.DataFrame(test, columns=c_names)

# The below line of code creates a logrithmically transformed target variable.
train_target_log = np.log(train_target)
```

## SVM 

### C hyperparamter
The first hyperparameter we will discuss is C. According to the documentation for the SVR function on scikit-learn, this is a regularization parameter [1]. When I say regularization, this is in regards to a tehnique used to mitigate potential overfitting. I of course have had many instances where my model has a somewhat decent mean absolute error, but when it was tested with new data, the mean absolute error skyrockets. What happened here was that the model pretty much focused insanely well on the training data, but when feed with unseen data, performed awful. For the SVR function, the strength of regularization will be inversely proportional to C [1]. In other words, lower values of C will result in stronger regularization (hopefully reducing overfitting).

```python
# The below lines of code create a train_df and test_df data frames to store C values and corresponding MAEs.
train_df = pd.DataFrame(columns=['C','mae'])
test_df = pd.DataFrame(columns=['C','mae'])

# The below line of code creates a for loop that goes from 0.5 to 20, increasing in 0.5 increments.
for i in np.arange(0.5, 20, 0.5):
    # The below line of code creates a support vector regressor with C equal to i.
    svr = SVR(C=i)
    # The below line of code fits the support vector regressor with the training data.
    svr.fit(train, train_target_log)
    # The below line of code predicts the target training values using the training predictor values.
    pred_target_log = svr.predict(train)
    # The below line of code reverses the log transformation to get the actual predictions.
    pred_target = np.exp(pred_target_log)
    # The below line of code calculates the mean absolute error.
    mae = mean_absolute_error(train_target, pred_target)
    # The below line of code updates the train_df data frame.
    train_df.loc[len(train_df.index)] = [i, mae]
    # The below line of code predicts the target test values using the testing predictor values.
    pred_target_log = svr.predict(test)
    # The below line of code reverses the log transformation to get the actual predictions.
    pred_target = np.exp(pred_target_log)
    # The below line of code calculates the mean absolute error.
    mae = mean_absolute_error(test_target, pred_target)
    # The below line of code updates the test_df data frame.
    test_df.loc[len(test_df.index)] = [i, mae]
```

```python
# The below lines of code plot the mae for the train_df as a dashed line while the mae for the test_df is a dotted line.
sns.lineplot(train_df['mae'], color='blue', linestyle='dashed')
sns.lineplot(test_df['mae'], color='blue', linestyle='dotted')
# The following line of code creates a title for the graph.
plt.title('train_df MAE (dashed) and test_df MAE (dotted)')
# The following line of code creates an x-axis label.
plt.xlabel('C')
plt.show()
```

The dashed line represents the mean absolute error (MAE) on the training data, while the dotted line represents the MAE on the testing/unseen data. On the left side of the graph, the two lines are closest to each other, but as we move to the right of the graph, the gap between them increases.
 
This behavior is because the x-axis represents different values of C. When C is small, the model applies a stronger regularization, resulting in a better generalization to unseen data. As C is increased, due to less regularization, the model tries harder to fit the training data perfectly, leading to overfitting. As we can see this lowers the MAE on the training data, it does increase the MAE on the testing/unseen.

Before the MAE for testing data increases, we can see there is an optimal value for C where the MAE (dotted line) decreases before it starts to increase again. Past this, increasing C leads to less regularization.

### Kernel Hyperparameter

Now, there are three different kernels that we could use: linear, RBF, and polynomial. While each is interesting to learn about, in the interest of time, I will provide a visual demonstration of what's happening with each.



The graph above comes from the 'Suport Vector Regression (SVR) using linear and non-linear kernels' page on scikit-learn [2]. As we can see, the linear kernel results in a straight regression line, which may not predict the target values well when the relationship between the features is non-linear. On the other hand, the polynomial and RBF models create non-linear regression lines that better fit the data points, allowing them to predict the target values more effectively.

Although the above example is two-dimensional, our models work in 30 plus dimensions, so a hyperplane would be used. However, the concepts remain the same.

For those interesting in learning more about the different types of kernels beyond what is described above, links will be provided.

```python
# The below line of code creates train_df and test_df data frames to store kernel types and corresponding MAEs.
train_df = pd.DataFrame(columns=['kernel', 'mae'])
test_df = pd.DataFrame(columns=['kernel', 'mae'])
# The below line of code creates a list to hold different kernel values.
kern = ['rbf', 'linear', 'poly']

# The following line of code creates a loop that iterates through each kernel type in kern.
for i in kern:
    # The below line of code creates a support vector regressor with the kernel specified by i.
    svr = SVR(kernel=i)
    # The following line of code fits the support vector regressor to the training data.
    svr.fit(train, train_target_log)
    # The below line of code predicts the target training values using the training predictor values.
    pred_target_log = svr.predict(train)
    # The following line of code reverses the log transformation to get the actual predictions.
    pred_target = np.exp(pred_target_log)
    # The below line of code calculates the mean absolute error.
    mae = mean_absolute_error(train_target, pred_target)
    # The following line of code updates the train_df data frame.
    train_df.loc[len(train_df.index)] = [i, mae]
    # The below line of code predicts the target test values using the testing predictor values.
    pred_target_log = svr.predict(test)
    # The following line of code reverses the log transformation to get the actual predictions.
    pred_target = np.exp(pred_target_log)
    # The below line of code calculates the mean absolute error.
    mae = mean_absolute_error(test_target, pred_target)
    # The following line of code updates the test_df data frame.
    test_df.loc[len(test_df.index)] = [i, mae]
```

```python
# The below line of code plots a barchart where each kernels training MAE is plotted.
sns.barplot(x=train_df['kernel'], y=train_df['mae'])
plt.show()
```

The graph above indicates that the Support Vector Machine (SVM) using the RBF kernel has the lowest mean absolute error (MAE), while the SVM with the linear kernel has the highest MAE. This suggests that our features may not have a linear relationship with the target variable, and the non-linear kernels (like RBF and polynomial) are better suited to capture the underlying patterns in the data.

```python
# The following line of code plots a barchart where each kernels testing MAE is plotted.
sns.barplot(x=test_df['kernel'], y=test_df['mae'])
plt.show()
```

The graph above indicates that the Support Vector Machine (SVM) with the RBF kernel has the lowest mean absolute error (MAE). Meanwhile, the SVM with the polynomial kernel performed worse on the testing data compared to the linear kernel, potentially indicating that the polynomial kernel led to overfitting on the training data.

While the polynomial and RBF kernels have additional hyperparameters that could be fine-tuned, I will not dive into them here due to time constraints. These adjustments will only be discussed further if either the polynomial or RBF models turn out to produce the best results.

### Epsilon

According to Géron in "Hands-On Machine Learning with Scikit-Learn & TensorFlow", the epsilon parameter controls the width of the 'street' or 'tube' within which the model tries to fit as many data points as possible while limited data points outside of the 'street' or 'tube' [3]. Higher values of epsilon will lead to a wider 'street' or 'tube' whiel smaller value of epsilon will lead to a narrower 'street' or 'tube'.

```python
# The below line of code creates train_df and test_df data frames to store epsilon values and corresponding MAEs.
train_df = pd.DataFrame(columns=['e', 'mae'])
test_df = pd.DataFrame(columns=['e', 'mae'])

# The following line of code creates a loop that iterates from 0.01 to 0.5 in increments of 0.01.
for i in np.arange(0.01, 0.5, .01):
    # The below line of code creates a support vector regressor with epsilon equal to i.
    svr = SVR(epsilon=i)
    # The following line of code fits the support vector regressor to the training data.
    svr.fit(train, train_target_log)
    # The below line of code predicts the target training values using the training predictor values.
    pred_target_log = svr.predict(train)
    # The following line of code reverses the log transformation to get the actual predictions.
    pred_target = np.exp(pred_target_log)
    # The below line of code calculates the mean absolute error.
    mae = mean_absolute_error(train_target, pred_target)
    # The following line of code updates the train_df data frame.
    train_df.loc[len(train_df.index)] = [i, mae]
    # The below line of code predicts the target test values using the testing predictor values.
    pred_target_log = svr.predict(test)
    # The following line of code reverses the log transformation to get the actual predictions.
    pred_target = np.exp(pred_target_log)
    # The below line of code calculates the mean absolute error.
    mae = mean_absolute_error(test_target, pred_target)
    # The following line of code updates the test_df data frame.
    test_df.loc[len(test_df.index)] = [i, mae]
```

```python
# The below lines of code plot the mae for the train_df as a dashed line while the mae for the test_df is a dotted line.
sns.lineplot(train_df['mae'], color='blue', linestyle='dashed')
sns.lineplot(test_df['mae'], color='blue', linestyle='dotted')
# The below line of code specifies x-axis represents different epsilon value.
plt.xlabel('Epsilon')
plt.show()
```

From the above graph, there is an issue with the scale on the x-axis (it should be in .1 increments, not 10). Ignoring that, we can see that the biggest difference between the dashed (training) and dotted (testing) mean absolute error is around epsilon equal to 0. As epsilon starts to increase, the MAE for the training data begins to increase while the MAE for the testing data does seem to bump up and down. As epsilon continues to increase, the training and testing MAE become indistinguishable from one another, indicating that at a certain point the model does the same job at predicting the training data and the testing data, but the MAE is much higher than when there was a difference.

## Finding the best combination of hyperparameters

Now that we know more about the three main hyperparameters, we can attempt to figure out what combination of the three works best.

```python
# Below line of creates creates a dataframe to hold the kernel, C, epsilon, and mae for each SVR model.
svr_df = pd.DataFrame(columns=['kernel', 'C', 'epsilon', 'mae'])

# Below line of code defines best_score function.
def best_score(c_vals, epsilo, kern):
    # Below three lines of code iterate through every entry in kern, c_vals, and epsilo.
    for i in kern:
        for c in c_vals:
            for e in epsilo:
                # The below lines of code fit an Support Vector Regressor, predicting target values, transform target values, calculate the MAE, and update the svr_df data frame with the parameters used and MAE.
                svr = SVR(kernel=i, epsilon=e, C=c)
                svr.fit(train, train_target_log)
                pred_target_log = svr.predict(test)
                pred_target = np.exp(pred_target_log) 
                mae = mean_absolute_error(test_target, pred_target)
                svr_df.loc[len(svr_df.index)] = [i, c, e, mae]

# The below three lines of code create list of values to be used to C, epsilon, and kernel parameters in SVR model building.
c_vals = np.arange(0.5, 10, .5).tolist()
epsilo = np.arange(.1, 1, .1).tolist()
kern = ['rbf','poly','linear']
# The below line of code calls the best_score function with the c_vals, epsilon, and kern list.
best_score(c_vals, epsilo, kern)

# The below line of code sorts through the svr_df data frame so the models with the 10 lowest MAEs are visible.
svr_df.sort_values(by='mae', ascending=True).head(10)
```

Since all of the 10 lowest MAEs used an RBF kernel, this will be the main kernel as we further explore the best values for C and epsilon. The optimal value for C is 3.0, and the optimal value for epsilon is 0.1. Next, we will use the best_score function with a range of values for C that includes 3 as a central point, along with 0.1 for epsilon, to see if there are any other values that result in lower MAEs.

```python
# The below lines of code specify different values to be used for C and epsilon.
c_vals = np.arange(2.0, 4.0, .1).tolist()
epsilo = np.arange(0, .20, .01).tolist()
kern = ['rbf']
# The following line of code calls the best_score function with c_vals, epsilo, and kern
best_score(c_vals, epsilo, kern)

# The following line of code sorts throught the values of the svr_df data frame to view the SVR models with the 10 lowest MAEs.
svr_df.sort_values(by='mae', ascending=True).head(10)
```

The SVR model with the lowest MAE has a RBF kernel, C value of 2.1, and epsilon value of 0.06.

```python
svr = SVR(kernel='rbf', epsilon=0.06, C=2.1)
svr.fit(train, train_target_log)
pred_target_log = svr.predict(train)
pred_target = np.exp(pred_target_log)
mae = mean_absolute_error(train_target, pred_target)
print('MAE on training data:', np.round(mae,2))
pred_target_log = svr.predict(test)
pred_target = np.exp(pred_target_log) 
mae = mean_absolute_error(test_target, pred_target)
print('MAE on testing data:', np.round(mae,2))
```

The SVR model has an MAE of 54,345 on the training data and an MAE of 71,541 on the testing data. This indicates that when fed unseen data, our model was off by around $71,541 in predicting the selling price of a single residential property in the Johnson County area.
