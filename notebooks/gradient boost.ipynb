```python
# The below lines of code import the pandas library and the display function.
import pandas as pd 
from IPython.display import display

# The below lines of code read in the train_df and test_df CSV files.
train = pd.read_csv('train_df.csv')
test = pd.read_csv('test_df.csv')

# The below lines of code display the first five rows of the train data frame.
print('First five rows of train data frame:')
display(train.head())


print()
print()
# The below lines of code display the first five rows of the test data frame.
print('First five rows of test data frame:')
display(test.head())

# The below lines of code delete the previous index column in both the train and test data frames. 
del train['Unnamed: 0']
del test['Unnamed: 0']


print()
print()
# The below lines of code display the first five rows of the train data frame.
print('First five rows of train data frame after removing unnecesary column:')
display(train.head())

print()
print()
# The below lines of code display the first five rows of the test data frame.
print('First five rows of test data frame after removing unnecesary column:')
display(test.head())


from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# The below lines of code assign the Price column from the train and test data frames as the target variables.
train_target = train['Price']
test_target = test['Price']

# The below lines of code delete the Price column from the train and test data frames. 
del train['Price']
del test['Price']

train_og = train.copy()
test_og = test.copy()
# The below lines of code create a transformation pipeline that applies a logarithmic transformation followed by a standard scaling.
pipe1 = Pipeline([
    ('log', FunctionTransformer(func=np.log1p)),
    ('scaler', StandardScaler()),
    ])

# The below line of code retrieves the column names from the train data frame.
c_names = train.columns


# The below line of code fits the pipe1 pipeline with the training data, and then transforms the train data with the fitted pipeline.
train = pipe1.fit_transform(train)
# The below line of code transforms the test data with the fitted pipeline.
test = pipe1.transform(test)

# The below lines of code create new train and test dataframes using the original column names. 
train = pd.DataFrame(train, columns=c_names)
test = pd.DataFrame(test, columns=c_names)

# The below line of code creates a logrithmically transformed target variable.
train_target_log = np.log(train_target)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.metrics import mean_absolute_error

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

## Fitting an inital gradient boosting regressor and using randomized search cv to find the parameters with the lowest error.

```python
# The following lines of code create a dictionary with different hyperparameter values.
param_grid = {
    'n_estimators': np.arange(1,200,5),
    'max_depth': np.arange(5,100,5),
    'min_samples_split': np.arange(2,20,1),
    'learning_rate': np.arange(.01,1,.01),
    'subsample': np.arange(.01,1,.01),
}

rfr = RandomForestRegressor(max_depth=45, max_features=21, min_samples_split=5,
                            n_estimators=60, random_state=42)
rfr.fit(train, train_target)

# The below line of code specifies an intial Random Forest Regressor.
gbr = GradientBoostingRegressor(random_state=42, init=rfr)

# The following initalizes a RandomizedSearchCV model with the Random Forest Regressor, using param_grid for hyparameters, neg_mean_absolute_error as the scoring metric, 4-fold cross-validation, and a random state for reproducibility.
grid_search = RandomizedSearchCV(gbr, 
                                 param_distributions=param_grid,
                                 scoring='neg_mean_absolute_error',
                                 cv=4,
                                 return_train_score=True,
                                 random_state=22,
                                 n_iter=130
                                 )

# The below line of code fits the grid_search with train and train_target.
grid_search.fit(train, train_target)
```

Using the best model found from above, testing it on our data.

```python
gbr = grid_search.best_estimator_
print(gbr)
print()

# The following lines of code fit a Random Forest Regressor and calculates the mean absolute the mean absolute error of the model on the training and testing data.
gbr.fit(train, train_target_log)
pred_target_log = gbr.predict(train)
pred_target = np.exp(pred_target_log)
print('Training data MAE:', np.round(mean_absolute_error(train_target, pred_target),2))
pred_target_log = gbr.predict(test)
pred_target = np.exp(pred_target_log)
print('Testing data MAE:', np.round(mean_absolute_error(test_target, pred_target),2))
```

Seeing that the difference between the training and testing data MAE is around 50,000 is not good, and suggests major overfitting.

## Tinkering around a little bit with the parameters

In hopes of limiting the overfitting

```python
gbr = GradientBoostingRegressor(init=RandomForestRegressor(max_depth=45,
                                                           max_features=21,
                                                           min_samples_split=5,
                                                           n_estimators=60,
                                                           random_state=42),
                          learning_rate=0.01, max_depth=15,
                          min_samples_split=20, n_estimators=186,
                          random_state=42, min_samples_leaf=10, 
                          subsample=0.7000000000000001)

# The following lines of code fit a Random Forest Regressor and calculates the mean absolute the mean absolute error of the model on the training and testing data.
gbr.fit(train, train_target_log)
pred_target_log = gbr.predict(train)
pred_target = np.exp(pred_target_log)
print('Training data MAE:', np.round(mean_absolute_error(train_target, pred_target),2))
pred_target_log = gbr.predict(test)
pred_target = np.exp(pred_target_log)
print('Testing data MAE:', np.round(mean_absolute_error(test_target, pred_target),2))
```

We were able to alieviate the overfitting a little bit, but there is still a difference of around 40,000 between the testing MAE and the training MAE. We did this by limiting the max_depth parameter.

## PCA

The next try is PCA (reducing the number of features.)

```python
from sklearn.decomposition import PCA
```

Going back to the inital notebook for this project, we dummy encoded all the values for City and Zip code. This of course created a great number of additional columns. While this may not be statistically sound, I am going to utilize PCA for these categorical columns in hopes of reducing the dimensionality, and hopefully a better model.

```python
pc = PCA(random_state=22, n_components=0.9)
train_cat_pca = pc.fit_transform(train.iloc[:,6:])
train_cat_pca = pd.DataFrame(train_cat_pca)
test_cat_pca = pc.transform(test.iloc[:,6:])
test_cat_pca = pd.DataFrame(test_cat_pca)

train_new = train.iloc[:,:6]
train_new = pd.concat([train_new, train_cat_pca], axis=1)
test_new = test.iloc[:,:6]
test_new = pd.concat([test_new, test_cat_pca], axis=1)

train_new.columns = train_new.columns.astype(str)
test_new.columns = test_new.columns.astype(str)
```

```python
gbr = GradientBoostingRegressor(init=RandomForestRegressor(max_depth=45,
                                                           max_features=21,
                                                           min_samples_split=5,
                                                           n_estimators=60,
                                                           random_state=42),
                          learning_rate=0.01, max_depth=13,
                          min_samples_split=20, n_estimators=186,
                          random_state=42, min_samples_leaf=10, 
                          subsample=0.7000000000000001, warm_start=False)


# The following lines of code fit a Random Forest Regressor and calculates the mean absolute the mean absolute error of the model on the training and testing data.
gbr.fit(train_new, train_target_log)
pred_target_log = gbr.predict(train_new)
pred_target = np.exp(pred_target_log)
print('Training data MAE:', np.round(mean_absolute_error(train_target, pred_target),2))
pred_target_log = gbr.predict(test_new)
pred_target = np.exp(pred_target_log)
print('Testing data MAE:', np.round(mean_absolute_error(test_target, pred_target),2))
```

As we can see, our MAE on our testing data did decrease, but there is still a difference of around 40,000 between the training and testing MAE.
